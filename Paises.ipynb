```python
import pandas as pd
import plotly.express as px
from IPython.display import display, HTML

# Criar DataFrame (simulando a planilha Excel)
dados = {
    "Pa√≠s": [
        "Brasil", "Estados Unidos", "China", "√çndia", "Alemanha",
        "Jap√£o", "Fran√ßa", "Canad√°", "Austr√°lia", "√Åfrica do Sul"
    ],
    "Continente": [
        "Am√©rica do Sul", "Am√©rica do Norte", "√Åsia", "√Åsia", "Europa",
        "√Åsia", "Europa", "Am√©rica do Norte", "Oceania", "√Åfrica"
    ],
    "Popula√ß√£o (milh√µes)": [214, 332, 1412, 1408, 83, 125, 65, 39, 26, 61],
    "PIB (US$ trilh√µes)": [2.3, 26.9, 17.7, 3.7, 4.4, 4.2, 3.1, 2.2, 1.8, 0.4],
    "Expectativa de Vida": [75, 78, 77, 70, 81, 84, 82, 82, 83, 64],
    "IDH": [0.758, 0.926, 0.768, 0.644, 0.942, 0.925, 0.901, 0.936, 0.951, 0.713],
}

df = pd.DataFrame(dados)

# ================================================
# 1Ô∏è‚É£ Gr√°fico ‚Äî Mapa de Popula√ß√£o
# ================================================
fig_mapa = px.choropleth(
    df,
    locations="Pa√≠s",
    locationmode="country names",
    color="Popula√ß√£o (milh√µes)",
    hover_name="Pa√≠s",
    title="üåç Popula√ß√£o por Pa√≠s (em milh√µes)",
    color_continuous_scale="viridis"
)
fig_mapa.show()

# ================================================
# 2Ô∏è‚É£ Gr√°fico ‚Äî Ranking de PIB
# ================================================
fig_pib = px.bar(
    df.sort_values("PIB (US$ trilh√µes)", ascending=False),
    x="Pa√≠s",
    y="PIB (US$ trilh√µes)",
    color="PIB (US$ trilh√µes)",
    title="üí∞ Ranking de PIB (em trilh√µes de US$)",
    text_auto=True,
    color_continuous_scale="plasma"
)
fig_pib.show()

# ================================================
# 3Ô∏è‚É£ Gr√°fico ‚Äî IDH x Expectativa de Vida
# ================================================
fig_idh = px.scatter(
    df,
    x="IDH",
    y="Expectativa de Vida",
    size="Popula√ß√£o (milh√µes)",
    color="Continente",
    hover_name="Pa√≠s",
    title="üìà IDH x Expectativa de Vida",
)
fig_idh.show()

# ================================================
# 4Ô∏è‚É£ Tabela de Ranking Populacional
# ================================================
df_rank = df.sort_values("Popula√ß√£o (milh√µes)", ascending=False).reset_index(drop=True)
df_rank.index += 1
display(HTML("<h3>üèÜ Ranking Populacional</h3>"))
display(df_rank[["Pa√≠s", "Popula√ß√£o (milh√µes)", "PIB (US$ trilh√µes)", "IDH"]])
```
